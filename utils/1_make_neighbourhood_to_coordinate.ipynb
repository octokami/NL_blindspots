```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.metrics.pairwise import haversine_distances
```

```python
df = pd.read_csv("postcode_coordinate.csv")
df
```

```python
df['lat'] = np.radians(df.latitude)
```

```python
df['lon'] = np.radians(df.longitude)
```

```python
df
```

```python
coords = df[['lat', 'lon']]
```

```python
coords
```

```python
distance_matrix = haversine_distances(coords, coords)
```

```python
distance_matrix_df = pd.DataFrame(
    distance_matrix,
    index=pd.MultiIndex.from_frame(coords),
    columns=pd.MultiIndex.from_frame(coords)
)

distance_matrix_df
```

```python
mean_distance = distance_matrix_df.apply(lambda x: np.mean(x) * 6371).rename('mean_distance')

mean_distance
```

```python
result = df.join(mean_distance, on=['lat', 'lon'])
result
```
